1. What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?

 **Type I Error**
- It occurs when the null hypothesis (H₀) is true, but we incorrectly reject it.
- α (alpha) represents the level ofsignificance.        
 **Type II Error**
- It occurs when the null hypothesis (H₀) is false, but we fail to reject it.
- β (beta) — related to the power of the test, where  Power = 1 - β.
 **Impact on Decision-Making.**
- **Type I Error:** Leads to false alarms. Resources might be wasted pursuing false discoveries (e.g., approving an ineffective drug).
- **Type II Error:** Leads to missed opportunities. We might overlook a real effect (e.g., rejecting a drug that could have saved lives).

2. What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?
- The P-value (probability value) is a key concept in hypothesis testing that helps you decide whether to reject or fail to reject the null hypothesis (H₀).
- It tells you how likely your sample data would be if H₀ were actually true.

| P-value        | Interpretation                                  | Decision about H₀                                                 |
| -------------- | ----------------------------------------------- | ----------------------------------------------------------------- |
| **Low (≤ α)**  | The observed data is **very unlikely** under H₀ | **Reject H₀** — evidence supports the alternative hypothesis (H₁) |
| **High (> α)** | The observed data is **likely** under H₀        | **Fail to reject H₀** — not enough evidence to support H₁        
 |
